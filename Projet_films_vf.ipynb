```python
from sklearn.manifold import TSNE
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import sys
sys.path.append('../')
import iads as iads
from iads import kmoyennes as km
from iads import utils as ut
from iads import LabeledSet as ls
from iads import Classifiers as cl
from iads import ad
from iads import ad1
%load_ext autoreload
%autoreload 2
```

```python
base=pd.read_csv("./genre_films.csv")
```

```python
base=base.set_index("title")
```

```python
classes=base["genre"].values
```

```python
for i in range(len(classes)):
    classes[i]=1 if classes[i]=="Hommes" else -1
        
```

```python
y=base.iloc[:,0:3]
x=base.iloc[:,0:3].values

normals=(x-y.mean().values)/y.std().values
```

```python
DataSet=ls.LabeledSet(3)
```

```python
x=base[["popularity","vote_count","budget"]].values
```

```python
for i in range(len(x)):
    DataSet.addExample(normals[i],classes[i])
```

### Application de l'algorithme des K plus proches voisins

```python
accuracy=list()
for i in range(10):
    setApp,setTest=ut.split(DataSet,90)
    KNN= cl.ClassifierKNN(3,11)
    KNN.train(setApp)
    accuracy.append(KNN.accuracy(setTest))
print(accuracy)
```

```python
print("Moyenne de taux de bonne classification = ",np.mean(accuracy))
print("Ecart type = ",np.std(accuracy))
```

```python
fig, ax = plt.subplots()
y_pos = np.arange(1,11)

plt.bar(y_pos, accuracy, align='center',
        color='blue', ecolor='black')

ax.set_xticks(y_pos)
ax.set_xticklabels(y_pos)
#ax.invert_yaxis()  # labels read top-to-bottom
ax.set_ylabel('accuracy')


plt.title("10 folds cross validation")
plt.show()
```

```python
plt.plot(np.arange(1,11),accuracy)
plt.xticks(np.arange(1,11))
plt.ylabel("Accuracy")
plt.xlabel("Folds")
plt.title("10 folds cross validation")
```

### Application des arbres de décision (approche out of bag)

```python
nom_num={"budget":0,"popularity":1,"vote_count":2}
```

```python
perfApp,perfTest=[],[]
for i in range(1,11):
    foret = ad.ClassifierBaggingTreeOOB(i,0.66,0.0,False,['budget','popularity','vote_count'],nom_num)
    foret.train(DataSet)
    couple=foret.accuracy(DataSet)
    perfApp.append(couple[0])
    perfTest.append(couple[1])
```

```python
ax =plt.subplot()
plt.rcdefaults()
ax.set_ylabel("Accuracy ")
ax.set_xlabel("Nb d'arbres construits")
ax.set_xticks(np.arange(1,11))
ax.plot(np.arange(1,11),perfApp,'b',np.arange(1,11),perfTest,'r')
ax.legend(["X=(apprentissage)","notX =(Test)"])
plt.title("Accuracy obtenu avec l'approche Out Of the Bag")

plt.show()
```

### Partie 03: prediction de la note moyenne

```python
from sklearn.manifold import TSNE
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import sys
sys.path.append('../')
import iads as iads
from iads import kmoyennes as km
from iads import utils as ut
from iads import LabeledSet as ls
from iads import Classifiers as cl
from iads import ad
from iads import ad1
%load_ext autoreload
%autoreload 2
```

```python
base =pd.read_csv("MyDataBase.csv")
base=base.set_index("title")
base=base.drop(["categorie"],axis=1)
```

```python
base=base.drop(["id","revenue","vote_count"],axis=1)
```

```python
notes=base["vote_average"].values
```

```python
y=base.iloc[:,0:4]
x=base.iloc[:,0:4].values

normals=(x-y.mean().values)/y.std().values
```

```python
DataSet=ls.LabeledSet(4)
```

```python
for i in range(len(x)):
    DataSet.addExample(normals[i],notes[i])
```

**Application de l'algorithme des k plus proches voisins**

```python
ecart1=list()
setApp,setTest=ut.split(DataSet,90)
for i in range(3,31):
    KNN= cl.ClassifierKNN_Reg(4,i)
    KNN.train(setApp)
    ecart1.append(KNN.ecart(setTest))
```

```python
plt.plot(np.arange(3,31),ecart1,color='orange')
plt.title("Ecart type en fonction de la valeur de k")
plt.xlabel("valeur de k")
plt.ylabel("Ecart type")
```

Meilleure valeur de K : k>20

```python
ecart=list()
for i in range(10):
    setApp,setTest=ut.split(DataSet,90) 
    KNN= cl.ClassifierKNN_Reg(4,50)
    KNN.train(setApp)
    ecart.append(KNN.ecart(setTest))
```

```python
plt.bar(np.arange(1,11),ecart,color='red')
plt.title("10 folds cross validation")
plt.xlabel("folds")
plt.xticks(np.arange(1,11))
plt.ylabel("Ecart moyen")
```
